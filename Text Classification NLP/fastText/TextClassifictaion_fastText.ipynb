```python
import pandas as pd

df = pd.read_csv('/content/ecommerceDataset.csv.zip',names = ['category','description'],header=None)
print(df.shape)
df.head(3)
```

```python
df.category.value_counts()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df.category.replace('Clothing & Accessories', 'Clothing_Accessories', inplace=True)
df.category.unique()
```

```python
df['category'] = "__label__" + df['category'].astype(str)
df.head(4)
```

```python
df['category_description'] = df['category'] + ' '+  df['description']
df.head(3)
```

```python
df['category_description'][0]
```

```python
import re

def preprocess(text):

  text = re.sub(r'[^\w\s\']', ' ', text)
  text = re.sub(r' +', ' ', text)
  return text.strip().lower()
```

```python
preprocess( "  RAJIII's |  Bookstationary / Dictionary shelf (collapsed/3/4th, black) | ? . ki")
```

```python
df['category_description'] = df['category_description'].map(preprocess)
```

```python
df.head(3)
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df,test_size = 0.2)
```

```python
train.shape,  test.shape
```

```python
train.head(3)
```

```python
test.head(3)
```

```python
train.to_csv('ecommerce.train', columns = ['category_description'], index = False, header = False)
test.to_csv('ecommerce.test', columns = ['category_description'], index = False, header = False)
```

```python
import fasttext

# Train on your training file
model = fasttext.train_supervised("ecommerce.train")

# Predict on a new text
labels, probs = model.predict("think and grow rich deluxe edition")
print(labels, probs)

# Evaluate on your test file
result = model.test("ecommerce.test")
print("Test samples:", result[0], "Precision:", result[1], "Recall:", result[2])
```
